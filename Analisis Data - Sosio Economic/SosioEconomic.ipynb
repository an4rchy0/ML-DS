```python
import pandas as pd

try :
    df = pd.read_csv('./Data/2021socio_economic_indonesia.csv')
    print(df)
except Exception as e:
    print(e)
```

Clear data

```python
missing_values = df.isnull().sum()
print(missing_values)

df = pd.DataFrame(df)
```

Menyortir Pendapatan GDP mulai dari terendah: ds1 (data sortir1)

```python
ds1 = df.sort_values(by='exp_percap')
ds1
```

Visualisasi Data Dengan Penduduk Paling Rendah

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
for pr, d in df.groupby('poorpeople_percentage'):
    ax.scatter(d['cities_reg'], d['poorpeople_percentage'], label=pr)

plt.title('Poor Region Rank')
plt.xlabel('Cities Region')
plt.ylabel('Poor Percentage (%)')
plt.grid(True)
ax.tick_params(axis='x', labelsize=10)
plt.xticks(rotation=90)
plt.show()
```

Sebaran Data Kota dengan Pendapatan

```python
import matplotlib.pyplot as plt;

fig, ax = plt.subplots()
for province, d in df.groupby('province'):
    ax.scatter(d['avg_schooltime'], d['exp_percap'], label=province)

plt.legend(loc='upper left', bbox_to_anchor=(1, 1))
plt.title('sebaran kota dengan exppercapita dan rata rata jam sekolah')
plt.xlabel('rata rata sekolah')
plt.ylabel('exp percapita')
plt.grid(True)
plt.show()
```
